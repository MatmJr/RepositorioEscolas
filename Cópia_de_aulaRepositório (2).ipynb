Vamos considerar que alto desempenho sejam notas maiores que 90. Quantos alunos ficaram com nota superior a 90 em redação? E qual o percentual?

```python
highRed = data[data['Nota_Redacao'] > 90]
```

```python
len(highRed)
```

```python
len(highRed) / len(estudantes) * 100
```

**25,47% das crianças tiraram nota maior que 90 em redação!**
